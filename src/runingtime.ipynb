```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
subtype = "H1"
matplotlib.rcParams['font.family'] = 'Times New Roman'
linewidth = 2.5
legend_fontsize = 17 # 图例字体大小
axis_fontsize = 19 # 横纵轴标题大小
title_fontsize = 21 # 标题字体大小
tick_fontsize = 17 # 刻度字体大小
markersize=15
dpi=300
figsize = (8, 6)

# 真实的 3 个采样点（数据点的位置）
x_points = [50824*0.333333, 50824*0.666666, 50284]

# 你希望显示的 7 个 x 轴刻度（刻度位置/标签）
x_ticks = [15000, 20000, 25000, 30000, 35000, 40000, 45000, 50000, 55000]

# 把“第一个数据点”的实际 x 也加成一个刻度，避免左侧起始刻度缺失看不出第一个点位置
first_point_tick = int(round(x_points[0]))
# x_ticks_all = sorted(set(x_ticks + [first_point_tick]))
x_ticks_all = x_ticks

methods = ['DTAN', 'DTAN-NoDAE']
num_methods = len(methods)

# 模拟数据（你需要换成真实数据）
train_time = [  [125.01, 259.41, 391.26],
                [298.05, 606.09, 984.72],
            ]

# colors = ['#7F83B5', '#A16C67', '#E4A06D', '#6C9567', '#E272E8', '#00B6C7', '#9467bd']
colors = ['#7F83B5', '#A16C67']
# markers = ['o', 'D', 's', '^', 'v', 'x', 'P']
markers = ['o', 'D']

plt.figure(figsize=figsize)

for i in range(num_methods):
    plt.plot(x_points, train_time[i], label=f'{methods[i]}', linestyle='-', marker=markers[i], markersize=markersize, color=colors[i], linewidth=linewidth)

# plt.yscale('log', base=2)
plt.ylabel("Training time (s)", fontsize=axis_fontsize)
plt.xlabel("Number of samples", fontsize=axis_fontsize)

# 显式控制 x 轴刻度（只传刻度位置，保持 ScalarFormatter，避免 ticklabel_format 报错）
plt.xticks(x_ticks_all, fontsize=tick_fontsize)
plt.yticks(fontsize=tick_fontsize)

# 让刻度范围完整可见（避免自动缩放导致部分刻度不显示）
plt.xlim(min(x_ticks_all) - 1000, max(x_ticks_all) + 1000)

all_values = [value for row in train_time for value in row]
plt.ylim(min(all_values)-200, max(all_values) * 1.3)
# plt.title("Training vs Inference Time")
plt.legend(ncol=1, fontsize=legend_fontsize, markerscale=0.5)
plt.grid(False)
plt.tight_layout()
plt.savefig(f"../result/figure/trainingtime_{subtype}.svg", dpi=dpi)
# plt.show()
```
